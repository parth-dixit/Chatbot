```python
import pandas as pd
import re
pd.set_option('display.max_colwidth',200)
```

```python
df = pd.read_csv(r'C:\Users\Parth Dixit\Downloads\Python_Real-World Machine Learning with Python\nscb.csv', encoding='latin-1')
df
```

```python
#Get data from 1st column and 0th index
convo = df.iloc[:,0]
convo
```

```python
#seperate the questions and answers from the dataset and storing them into list

clist = []
def qa_pairs(x):
    cpairs = re.findall(r": (.*?)(?:\.|\\n)", x)
    clist.extend(list(zip(cpairs[0:1], cpairs[1:])))
convo.map(qa_pairs);
convo_frame = pd.Series(dict(clist)).to_frame().reset_index()
convo_frame.columns = ['q', 'a']
#clist
```

```python
convo_frame
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
vectorizer = TfidfVectorizer(ngram_range=(1,3))
vec = vectorizer.fit_transform(convo_frame['q'])
```

```python
my_q = vectorizer.transform(['I cannot argue with you'])
cs = cosine_similarity(my_q, vec)
rs = pd.Series(cs[0]).sort_values(ascending=False)
top5 = rs.iloc[0:5]
top5
```

```python
def get_response(q):
    my_q = vectorizer.transform([q])
    cs = cosine_similarity(my_q, vec)
    rs = pd.Series(cs[0]).sort_values(ascending=False)
    rsi = rs.index[0]
    return convo_frame.iloc[rsi]['a']
```
